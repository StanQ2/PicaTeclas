```python
import pandas as pd

videogames = pd.read_csv('data.csv')
videogames
```

```python
def recorrer(n):
    arreglo = []
    for i in range(n):
        for j in range(i+1, n):
            if(videogames.iloc[i,6]==videogames.iloc[j, 6]):
                id1 = videogames.iloc[i,0] 
                id2 = videogames.iloc[j,0]
                arreglo.append((id1,id2))
    return arreglo
```

```python
def escribir_archivo(arreglo):
    with open('edges.csv', 'w') as file:
        for item in arreglo:
            file.write(f"{item[0]} , {item[1]}\n")

            
arreglo = recorrer(1499)
escribir_archivo(arreglo)
```
